```r
library(tm)
# install.packages("corpus")
library(corpus)
# install.packages("rvest")
library(rvest)
# install.packages("quanteda")
library(stringr)

data<-read.csv("nytarticles.csv", encoding = "latin1")

print("hi")
myCorpus <- VCorpus(VectorSource(data$text))

myCorpus <- tm_map(myCorpus, content_transformer(tolower))



remove1 <- function(x) gsub("http[[:alnum:][:punct:]]*", "", x)
remove2 <- function(x) gsub("http[[:alnum:]]*",'', x)
remove3 <- function(x)gsub('http\\S+\\s*', '', x) ## Remove URLs
remove4 <- function(x)gsub('\\b+RT', '', x) ## Remove RT
remove5 <- function(x)gsub('#\\S+', '', x) ## Remove Hashtags
remove6 <- function(x)gsub('@\\S+', '', x) ## Remove Mentions
remove7 <- function(x)gsub('[[:cntrl:]]', '', x) ## Remove Controls and special characters
remove8 <- function(x)gsub("\\d", '', x) ## Remove Controls and special characters
remove9 <- function(x)gsub('[[:punct:]]', '', x) ## Remove Punctuations
remove10 <- function(x)gsub("^[[:space:]]*","",x) ## Remove leading whitespaces
remove11<- function(x)gsub("[[:space:]]*$","",x) ## Remove trailing whitespaces
remove12 <- function(x)gsub(' +',' ',x) ## Remove extra whitespaces

remove13 <- function(x) gsub("[^\x01-\x7F]", "", x)

myCorpus <- tm_map(myCorpus,remove1)
myCorpus <- tm_map(myCorpus,remove2)
myCorpus <- tm_map(myCorpus,remove3)
myCorpus <- tm_map(myCorpus,remove4)
myCorpus <- tm_map(myCorpus,remove5)
myCorpus <- tm_map(myCorpus,remove6)
myCorpus <- tm_map(myCorpus,remove7)
myCorpus <- tm_map(myCorpus,remove8)
myCorpus <- tm_map(myCorpus,remove9)
myCorpus <- tm_map(myCorpus,remove10)
myCorpus <- tm_map(myCorpus,remove11)
myCorpus <- tm_map(myCorpus,remove12)
myCorpus <- tm_map(myCorpus,remove13)


myCorpus <- tm_map(myCorpus, stripWhitespace)
myCorpus <- tm_map(myCorpus, removePunctuation)
myCorpus <- tm_map(myCorpus, removeNumbers)


myCorpus <- tm_map(myCorpus, removeWords, stopwords('english'))
myCorpus <- tm_map(myCorpus, PlainTextDocument)

myCorpus <- tm_map(myCorpus, stemDocument, language = "english")


df <- data.frame(text = sapply(myCorpus, as.character), stringsAsFactors = FALSE)
# write.table(df,"ccscrub.txt",sep="\t",quote = FALSE, row.names=FALSE)
write.csv(df, file = "NYTSCRUB.csv",row.names=FALSE)

print("done")

```
